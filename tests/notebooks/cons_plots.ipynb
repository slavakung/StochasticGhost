```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
DATASET = 'compas'
sensitive_0_val = 'male'
sensitive_1_val = 'female'
```

```python
obj = pd.read_csv(str(DATASET)+'_iters/'+str(DATASET)+'_ftrial_new.csv').values
cons1 = pd.read_csv(str(DATASET)+'_iters/'+str(DATASET)+'_ctrial1_new.csv').values
cons2 = pd.read_csv(str(DATASET)+'_iters/'+str(DATASET)+'_ctrial2_new.csv').values
```

```python
maxiter = len(obj)
```

```python
for i in range(0, len(obj)):
    obj[i, :] = np.sort(obj[i, :])
    cons1[i, :] = np.sort(cons1[i, :])
    cons2[i, :] = np.sort(cons2[i, :])
```

```python
plt.plot(range(0, maxiter), cons1[:, 10], 'k-')
plt.fill_between(range(0, maxiter), cons1[:, 5], cons1[:, 15])
#plt.axhline(y=-0.001, color='black', linestyle='--', label='Horizontal Line at y=3')
plt.axhline(y=0.01, color='black', linestyle='--', label='Horizontal Line at y=3')
# plt.plot(range(0, maxiter+1), convals[:, 0], 'k-')
# plt.fill_between(range(0, maxiter+1), convals[:, 0], convals[:, 1])
plt.title("Constraint variation")
plt.xlabel('Iteration')
plt.ylabel('Difference in Avg loss over '+str(sensitive_1_val)+' and'+ str(sensitive_0_val))
#plt.yscale('log')
plt.plot()
```

```python
plt.plot(range(0, maxiter), cons2[:, 10], 'k-')
plt.fill_between(range(0, maxiter), cons2[:, 5], cons2[:, 15])
#plt.axhline(y=-0.001, color='black', linestyle='--', label='Horizontal Line at y=3')
plt.axhline(y=-0.01, color='black', linestyle='--', label='Horizontal Line at y=3')
# plt.plot(range(0, maxiter+1), convals[:, 0], 'k-')
# plt.fill_between(range(0, maxiter+1), convals[:, 0], convals[:, 1])
plt.title("Constraint variation")
plt.xlabel('Iteration')
plt.ylabel('Difference in Avg loss over '+str(sensitive_0_val)+' and'+ str(sensitive_1_val))
#plt.yscale('log')
plt.plot()
```

```python
cons = (cons1+cons2)/2
```

```python
plt.plot(range(0, maxiter), cons[:, 10], 'k-')
plt.fill_between(range(0, maxiter), cons[:, 5], cons[:, 15])
plt.axhline(y=-0.01, color='black', linestyle='--', label='Horizontal Line at y=3')
plt.axhline(y=0.01, color='black', linestyle='--', label='Horizontal Line at y=3')
# plt.plot(range(0, maxiter+1), convals[:, 0], 'k-')
# plt.fill_between(range(0, maxiter+1), convals[:, 0], convals[:, 1])
plt.title("Constraint variation")
plt.xlabel('Iteration')
plt.ylabel('|'+str(sensitive_0_val)+' - '+str(sensitive_1_val)+'|')
#plt.yscale('log')
plt.plot()
```
